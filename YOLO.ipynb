```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load the YOLOv8 model for pose estimation
model = YOLO('yolov8n-pose.pt')  # Replace with your specific YOLOv8 model if needed

# Function to determine if hands are raised
def is_hand_raised(keypoints):
    try:
        left_shoulder = keypoints[5]
        left_wrist = keypoints[7]
        right_shoulder = keypoints[6]
        right_wrist = keypoints[8]
        #print(left_shoulder)
        if (((left_wrist[1] < left_shoulder[1]) or (right_wrist[1] < right_shoulder[1])) and (((left_wrist[1]>0) and (left_shoulder[1]>0)) or ((right_wrist[1]>0) and (right_shoulder[1]>0)))):
            return True
    except IndexError:
        pass
        #print("galat hay vai")
    return False

# Open video capture
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform pose estimation using YOLOv8
    results = model(frame)
    
    

    for result in results:
        tw_frame= result.plot()
        #an_frame=result.copy()
        detections = result.keypoints.numpy() if result and result.keypoints is not None else []
        #print(detections)
        #keypoints = detections[:, :2]  # Extract x, y coordinates of keypoints
        #print(type(keypoints))
        #print(detections.data[0])
        # Check if hands are raised
        if is_hand_raised(detections.data[0]):
            cv2.putText(tw_frame, "Hands Raised", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
            #print("Yesssssssssssssssssssssss")
        else:
           cv2.putText(tw_frame, "Hands Not Raised", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
            #print("NOOOOOOOOOOOOOOOOOOOOOOOOOOOO")
    # Display the frame
    cv2.imshow("Hand Raise Detection", tw_frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()

```
